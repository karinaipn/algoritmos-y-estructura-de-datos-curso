```python
lista_=[1,5,2,4,6,3]

##### Worst case
### Término dominante O(n^2)
def BubleSort(lista_):
  for steps in range(len(lista_)-1,0,-1):
     print('steps',steps)
     for i in range(steps):#(6-1)
         print('indice:',i,"Me estas comparando con:",i+1)
         if lista_[i]>lista_[i+1]:
           temp=lista_[i]
           lista_[i]=lista_[i+1]
           lista_[i+1]=temp
           #print(lista_)     
  return lista_
```

```python
BubleSort([7,3,1,4,3,8,95])
```

```python
def unsortedList(size_of_array):
  import random

  random_list=[]
  for i in range(0,size_of_array):
    n=random.randint(1,100)
    random_list.append(n)
  return random_list
```

```python
BubleSort(unsortedList(10000))
```

```python
def BubleSortOptimized(lista): ##este algoritmo su mejor caso es cuando se le presenta una lista ordenada
                               ##Término dominante es O(n)

                               ###########Con un casi promedio, que sería una lista semi ordenada, O(n^2)
  swapped=True
  steps_=len(lista)-1
  while steps_>0 and swapped:
    print('steps:',steps_)
    swapped=False
    for i in range(steps_):
      print('indice:',i,'comparado con:',i+1)
      if lista[i]>lista[i+1]:
         swapped=True
         temp=lista[i]
         lista[i]=lista[i+1]
         lista[i+1]=temp
    steps_=steps_ - 1
    print(lista)
  return lista
```
