**Explica brevemente en una serie de pasos como trabaja el segundo algoritmo, pon ejemplos con listas o imágenes si lo requieres.**



1.   Se define la función que corresponde a datos de una lista
2.   Se crea una variable para guardar el valor correspondiente al indice 0 de la lista (set_min)
3. El ciclo for va a comparar cada posicion en la lista con la variable definida anteriormente, en caso de que sea menor sustituira al valor set_min
4. Finalmente al correr todos las posiciones arroja el valor minimo encontrado.

Minimos.jpg



```python
def findMinOptimizado(lista_):
  ##Best case
  SoloNumeros=[]
  for index in range (0,len(lista_)):
    if type(lista_[index])==int:
      SoloNumeros.append(lista_[index])

  set_min = SoloNumeros[0]
  for i in SoloNumeros: ##----------> O(n)
    if i < set_min:
      set_min = i
  return set_min
```
