```python
def linearSearch(items,value):
  for i in range(0,len(items)):  ###Término dominante en 
    if items[i]==value:
      return i
  return -1  #en caso de que no hubiera valor
```

```python
items=[2,4,6,34,25,4]
linearSearch(items,56)
```

**BINARY SEARCH**

```python
def binarySearch(items,value): #la funcion devuelve el indice en el que se encuentra el valor buscado
  low=0
  high=len(items)-1

  while low <= high:
    print('valor low:',low)
    print('valor high:',high)
    mid=(high+low)//2   #indice de la mitad de la lista, la doble linea es para que nos de un entero
    print('valor mind:',mid)
    if items[mid]< value:
      low=mid+1
    elif items[mid]>value:
      high=mid-1
    else:
      return mid
  return -1  #si no se encuentra el valor
```

```python
lista_=[2,4,8,12,16,24,36]
binarySearch(lista_,2)
```

```python
lista_2=[2,4,128,124,8,12,16,24,36] #no funciona para listas desordenadas
binarySearch(lista_2,124)
```

```python
lista_3=[2,4,6,6,16,24,36] #en listas con numeros repetidos, solo muestra el  primero
binarySearch(lista_3,6)
```

```python
def firstElement(items,value): #en listas con numeros repetidos
  low=0
  high=len(items)-1
  res=-1

  while low<=high:
    mid=(high+low)//2
    if items[mid]< value:
      low=mid+1
    elif items[mid]>value:
      high=mid-1
    else:
      res=mid
      high=mid-1
  return res 



def lastElement(items,value):
  low=0
  high=len(items)-1
  res=-1

  while low<=high:
    mid=(high+low)//2
    if items[mid]< value:
      low=mid+1
    elif items[mid]>value:
      high=mid-1
    else:
      res=mid
      low=mid+1
  return res 
```

```python
items=[2,3,4,5,5,5,5,5,5,56,7,8,12]

print("El primer elemento es:",firstElement(items,5))
print("El segundo elemento es:",lastElement(items,5)) 
```
