```python
def lista_anidada(seq, num):
    avg = len(seq) / float(num)
    out = []
    last = 0.0

    while last < len(seq):
        out.append(seq[int(last):int(last + avg)])
        last += avg

    return out
```

```python
import time

inicio=time.time()
lista_anidada_1=lista_anidada(range(100000000),10) #--------1

for i in range(0,len(lista_anidada_1)):#-------n
  for j in range(0,len(lista_anidada_1[0])): #-------A*B
    if lista_anidada_1[i][j]==4:#---------A*B
      print('El valor cuatro se encuentra')#---A*B

fin=time.time()
print(fin-inicio)
```

**Problema:**  Se tiene una lista de datos desordenada, requerimos encontrar el valor minímo de toda nuestra lista de datos.

Implemente una solución que sea óptima para asegurar que nuestros recursos satisfagan nuestra necesidad.

```python
lista=[5,4,2,1,1]

def findMin(alist):  #es para definir una función (Toda la parte de codigo que se escribio es la función)
  minimo=alist[0]
  for i in alist:
    valor_minimo=True
    for j in alist:
      if i>j:
        valor_minimo=False
    if valor_minimo:
      minimo=i
  return minimo
```

```python
print(findMin(lista))
```

```python
def findMax(alist):  #es para definir una función (Toda la parte de codigo que se escribio es la función)
  maximo=alist[0]
  for k in alist:
    valor_maximo=True
    for l in alist:
      if k<l:
        valor_maximo=False
    if valor_maximo:
      maximo=k
  return maximo
```

```python
print(findMax(lista))
```

Algunos otros casos que pudieran ocurrir 

```python
def findMinOptimizado(lista_):
  ##Best case
  set_min=lista_[0]
  for i in lista_: ##--------------->o(n)
    if i<set_min:
      set_min=i
  return set_min
```

```python
print(findMinOptimizado([5,4,2,1]))
```
